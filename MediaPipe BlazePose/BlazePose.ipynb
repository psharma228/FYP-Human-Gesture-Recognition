```
import cv2
from google.colab.patches import cv2_imshow
import numpy as np

# Read images with OpenCV.
images = {name: cv2.imread(name) for name in uploaded.keys()}

# Preview the images.
for name, image in images.items():
  print(name)   
  cv2_imshow(image)
```

All MediaPipe Solutions Python API examples are under mp.solutions.

For the MediaPipe Pose solution, we can access this module as `mp_pose = mp.solutions.pose`.

You may change the parameters, such as `static_image_mode` and `min_detection_confidence`, during the initialization. Run `help(mp_pose.Pose)` to get more informations about the parameters.

```
import mediapipe as mp

from sklearn.preprocessing import MinMaxScaler
mp_pose = mp.solutions.pose
pose = mp_pose.Pose(
    static_image_mode=False,upper_body_only=True, min_detection_confidence=0.5)

scaler = MinMaxScaler()
```

```
# Prepare DrawingSpec for drawing the face landmarks later.
mp_drawing = mp.solutions.drawing_utils 
drawing_spec = mp_drawing.DrawingSpec(thickness=1, circle_radius=1)
```

```

for name, image in images.items():
  results = pose.process(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
  image_hight, image_width, _ = image.shape
  if not results.pose_landmarks:
    continue
  print(
    f'Nose coordinates: ('
    f'{results.pose_landmarks.landmark[mp_pose.PoseLandmark.NOSE].x * image_width}, '
    f'{results.pose_landmarks.landmark[mp_pose.PoseLandmark.NOSE].y * image_hight}) \n'
    f'LEFT EYE Inner coordinates: ('
    f'{results.pose_landmarks.landmark[mp_pose.PoseLandmark.LEFT_EYE_INNER].x * image_width}, '
    f'{results.pose_landmarks.landmark[mp_pose.PoseLandmark.LEFT_EYE_INNER].y * image_hight})\n'
    f'LEFT EYE coordinates: ('
    f'{results.pose_landmarks.landmark[mp_pose.PoseLandmark.LEFT_EYE].x * image_width}, '
    f'{results.pose_landmarks.landmark[mp_pose.PoseLandmark.LEFT_EYE].y * image_hight})\n'
    f'LEFT EYE Outer coordinates: ('
    f'{results.pose_landmarks.landmark[mp_pose.PoseLandmark.LEFT_EYE_OUTER].x * image_width}, '
    f'{results.pose_landmarks.landmark[mp_pose.PoseLandmark.LEFT_EYE_OUTER].y * image_hight})\n'
    f'Right EYE Inner coordinates: ('
    f'{results.pose_landmarks.landmark[mp_pose.PoseLandmark.RIGHT_EYE_INNER].x * image_width}, '
    f'{results.pose_landmarks.landmark[mp_pose.PoseLandmark.RIGHT_EYE_INNER].y * image_hight})\n'
    f'Right EYE coordinates: ('
    f'{results.pose_landmarks.landmark[mp_pose.PoseLandmark.RIGHT_EYE].x * image_width}, '
    f'{results.pose_landmarks.landmark[mp_pose.PoseLandmark.RIGHT_EYE].y * image_hight})\n'
    f'Right EYE Outer coordinates: ('
    f'{results.pose_landmarks.landmark[mp_pose.PoseLandmark.RIGHT_EYE_OUTER].x * image_width}, '
    f'{results.pose_landmarks.landmark[mp_pose.PoseLandmark.RIGHT_EYE_OUTER].y * image_hight})\n'
    f'Left Ear coordinates: ('
    f'{results.pose_landmarks.landmark[mp_pose.PoseLandmark.LEFT_EAR].x * image_width}, '
    f'{results.pose_landmarks.landmark[mp_pose.PoseLandmark.LEFT_EAR].y * image_hight})\n'
    f'Right Ear coordinates: ('
    f'{results.pose_landmarks.landmark[mp_pose.PoseLandmark.RIGHT_EAR].x * image_width}, '
    f'{results.pose_landmarks.landmark[mp_pose.PoseLandmark.RIGHT_EAR].y * image_hight})\n'
    f'Mouth Left coordinates: ('
    f'{results.pose_landmarks.landmark[mp_pose.PoseLandmark.MOUTH_LEFT].x * image_width}, '
    f'{results.pose_landmarks.landmark[mp_pose.PoseLandmark.MOUTH_LEFT].y * image_hight})\n'
    f'Mouth Right coordinates: ('
    f'{results.pose_landmarks.landmark[mp_pose.PoseLandmark.MOUTH_RIGHT].x * image_width}, '
    f'{results.pose_landmarks.landmark[mp_pose.PoseLandmark.MOUTH_RIGHT].y * image_hight})\n'
    f'Left Shoulder coordinates: ('
    f'{results.pose_landmarks.landmark[mp_pose.PoseLandmark.LEFT_SHOULDER].x * image_width}, '
    f'{results.pose_landmarks.landmark[mp_pose.PoseLandmark.LEFT_SHOULDER].y * image_hight})\n'
    f'Right Shoulder coordinates: ('
    f'{results.pose_landmarks.landmark[mp_pose.PoseLandmark.RIGHT_SHOULDER].x * image_width}, '
    f'{results.pose_landmarks.landmark[mp_pose.PoseLandmark.RIGHT_SHOULDER].y * image_hight})\n'
    f'Left Elbow coordinates: ('
    f'{results.pose_landmarks.landmark[mp_pose.PoseLandmark.LEFT_ELBOW].x * image_width}, '
    f'{results.pose_landmarks.landmark[mp_pose.PoseLandmark.LEFT_ELBOW].y * image_hight})\n'
    f'Right Elbow coordinates: ('
    f'{results.pose_landmarks.landmark[mp_pose.PoseLandmark.RIGHT_ELBOW].x * image_width}, '
    f'{results.pose_landmarks.landmark[mp_pose.PoseLandmark.RIGHT_ELBOW].y * image_hight})\n'
    f'Left Wrist coordinates: ('
    f'{results.pose_landmarks.landmark[mp_pose.PoseLandmark.LEFT_WRIST].x * image_width}, '
    f'{results.pose_landmarks.landmark[mp_pose.PoseLandmark.LEFT_WRIST].y * image_hight})\n'
    f'Right Wrist coordinates: ('
    f'{results.pose_landmarks.landmark[mp_pose.PoseLandmark.RIGHT_WRIST].x * image_width}, '
    f'{results.pose_landmarks.landmark[mp_pose.PoseLandmark.RIGHT_WRIST].y * image_hight})\n'
     f'Left Pinky coordinates: ('
    f'{results.pose_landmarks.landmark[mp_pose.PoseLandmark.LEFT_PINKY].x * image_width}, '
    f'{results.pose_landmarks.landmark[mp_pose.PoseLandmark.LEFT_PINKY].y * image_hight})\n'
    f'Right Pinky coordinates: ('
    f'{results.pose_landmarks.landmark[mp_pose.PoseLandmark.RIGHT_PINKY].x * image_width}, '
    f'{results.pose_landmarks.landmark[mp_pose.PoseLandmark.RIGHT_PINKY].y * image_hight})\n'
    f'Left Index coordinates: ('
    f'{results.pose_landmarks.landmark[mp_pose.PoseLandmark.LEFT_INDEX].x * image_width}, '
    f'{results.pose_landmarks.landmark[mp_pose.PoseLandmark.LEFT_INDEX].y * image_hight})\n'
    f'Right Index coordinates: ('
    f'{results.pose_landmarks.landmark[mp_pose.PoseLandmark.RIGHT_INDEX].x * image_width}, '
    f'{results.pose_landmarks.landmark[mp_pose.PoseLandmark.RIGHT_INDEX].y * image_hight})\n'
    f'Left Thumb coordinates: ('
    f'{results.pose_landmarks.landmark[mp_pose.PoseLandmark.LEFT_THUMB].x * image_width}, '
    f'{results.pose_landmarks.landmark[mp_pose.PoseLandmark.LEFT_THUMB].y * image_hight})\n'
    f'Right Thumb coordinates: ('
    f'{results.pose_landmarks.landmark[mp_pose.PoseLandmark.RIGHT_THUMB].x * image_width}, '
    f'{results.pose_landmarks.landmark[mp_pose.PoseLandmark.RIGHT_THUMB].y * image_hight})\n'
    f'Left Hip coordinates: ('
    f'{results.pose_landmarks.landmark[mp_pose.PoseLandmark.LEFT_HIP].x * image_width}, '
    f'{results.pose_landmarks.landmark[mp_pose.PoseLandmark.LEFT_HIP].y * image_hight})\n'
    f'Right Hip coordinates: ('
    f'{results.pose_landmarks.landmark[mp_pose.PoseLandmark.RIGHT_HIP].x * image_width}, '
    f'{results.pose_landmarks.landmark[mp_pose.PoseLandmark.RIGHT_HIP].y * image_hight})\n'
  )
  # Draw pose landmarks.
  print(f'Pose landmarks of {name}:')
  annotated_image = image.copy()
  mp_drawing.draw_landmarks(
      image=annotated_image,
      landmark_list=results.pose_landmarks,
      landmark_drawing_spec=drawing_spec,
      connections=mp_pose.POSE_CONNECTIONS_UPPER,
      connection_drawing_spec=drawing_spec)
  cv2_imshow(annotated_image)

```
